<b> 1. Impact of Missing Data >>> Example: Impact on Salary Analysis

```python
# Create two dummy DataFrames: one with full data and one with missing values
full_data = {
    'Experience Level': ['Entry-level', 'Mid-level', 'Senior-level'],
    'Salary (USD)': [70000, 90000, 110000]
}
df_full = pd.DataFrame(full_data)

partial_data = {
    'Experience Level': ['Entry-level', 'Mid-level', 'Senior-level'],
    'Salary (USD)': [70000, None, 110000]
}
df_partial = pd.DataFrame(partial_data)
print("Full Data:\n", df_full)
print("\nPartial Data:\n", df_partial)
```

<b> 2) Identifying Missing Values

```python
# Count missing values in each column
missing_values = df.isna().sum()
print("Missing Values:\n", missing_values)
```

<b> 3) Strategies for Handling Missing Data >>> Dropping Missing Values

```python
# Calculate threshold for dropping missing values
threshold = len(df) * 0.05

# Filter columns with missing values below the threshold
cols_to_drop = missing_values[missing_values <= threshold].index

# Drop observations with missing values
df_cleaned = df.dropna(subset=cols_to_drop)
print("DataFrame after dropping missing values:\n", df_cleaned)
```

<b> 3) Strategies for Handling Missing Data >>> Imputing a Summary Statistic

```python
# Impute missing values with median salary
median_salary = df['Salary (USD)'].median()
df['Salary (USD)'].fillna(median_salary, inplace=True)
print("DataFrame after imputing missing values:\n", df)
```

<b> 4) Checking Remaining Missing Values

```python
# Check for remaining missing values
remaining_missing = df.isna().sum()
print("Remaining Missing Values:\n", remaining_missing)
```

<b> 5) Imputing by Sub-group

```python
# Impute missing values by sub-group (experience level)
median_salary_by_experience = df.groupby('Experience Level')['Salary (USD)'].median().to_dict()
df['Salary (USD)'] = df['Salary (USD)'].fillna(df['Experience Level'].map(median_salary_by_experience))
print("DataFrame after imputing missing values by sub-group:\n", df)
```

<b> 6) Confirmation of No Missing Values

```python
# Recheck for missing values
final_missing_values = df.isna().sum()
print("Final Missing Values:\n", final_missing_values)
```

<b> <font size = 12, font color ="yellow"> 7. Navigating Outliers: Strategies for Effective Data Analysis

<b> 1) Introduction to Outliers

```python
import pandas as pd

# Dummy dataset of house prices
house_prices = [250000, 300000, 400000, 450000, 500000, 5200000]

# Creating a DataFrame
df = pd.DataFrame({'House Prices': house_prices})

# Displaying summary statistics
print(df.describe())
```

<b> 2) Using Descriptive Statistics

```python
# Displaying mean, median, and standard deviation
print("Mean:", df['House Prices'].mean())
print("Median:", df['House Prices'].median())
print("Standard Deviation:", df['House Prices'].std())
```

<b> 3.1) Interquartile Range (IQR)

```python
# Calculating IQR
Q1 = df['House Prices'].quantile(0.25)
Q3 = df['House Prices'].quantile(0.75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

<b> 3.2) IQR in box plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a box plot
sns.boxplot(x=df['House Prices'])

# Adding labels
plt.xlabel('House Prices')
plt.title('Box Plot of House Prices')
plt.show()
```

<b> 3.3) Identifying thresholds

```python
# Calculating lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

<b> 4) Identifying Outliers

```python
# Identifying outliers
outliers = df[(df['House Prices'] < lower_bound) | (df['House Prices'] > upper_bound)]

print("Outliers:")
print(outliers)
```

<b> 5) Strategies for Handling Outliers

```python
# Removing outliers
cleaned_data = df[(df['House Prices'] >= lower_bound) & (df['House Prices'] <= upper_bound)]

print("Cleaned Data:")
print(cleaned_data)
```

<b> 6) Practice and Application

```python
# Creating histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['House Prices'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of House Prices (with Outliers)')
plt.xlabel('House Prices')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(cleaned_data['House Prices'], bins=10, color='salmon', edgecolor='black')
plt.title('Histogram of House Prices (without Outliers)')
plt.xlabel('House Prices')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import random

# Define the list of car makes, model years, and fuel types
car_makes = ['Toyota', 'Honda', 'Ford', 'Chevrolet', 'Tesla']
model_years = list(range(2010, 2023))
fuel_types = ['Gasoline', 'Electric', 'Hybrid']

# Generate dummy data
num_cars = 1000
cars = []
for i in range(num_cars):
    make = random.choice(car_makes)
    model_year = random.choice(model_years)
    fuel_type = random.choice(fuel_types)
    cars.append([make, model_year, fuel_type])

# Create a DataFrame from the generated data
columns = ['Make', 'Model_Year', 'Fuel_Type']
car_df = pd.DataFrame(cars, columns=columns)

# Analyze the distribution of categorical classes (fuel types)
fuel_type_counts = car_df['Fuel_Type'].value_counts()
print(fuel_type_counts)
```

```python
import pandas as pd
import numpy as np
import random

# Define the binary variable indicating fraudulent or non-fraudulent transactions
fraudulent_labels = ['Fraudulent', 'Non-Fraudulent']

# Generate dummy data
num_transactions = 1000
transactions = []
for i in range(num_transactions):
    label = random.choice(fraudulent_labels)
    transactions.append(label)

# Create a DataFrame from the generated data
transactions_df = pd.DataFrame(transactions, columns=['Fraudulent'])

# Check for class imbalance in binary categorical variables
fraudulent_counts = transactions_df['Fraudulent'].value_counts()
print(fraudulent_counts)
```

<b> 5. Class Frequency

```python
import pandas as pd
import numpy as np

# Create dummy data for credit card transactions
data = {
    'Transaction_Type': np.random.choice(['Purchase', 'Withdrawal', 'Transfer'], size=1000),
    'Customer_Segment': np.random.choice(['Segment A', 'Segment B', 'Segment C'], size=1000),
    'Purchase_Amount': np.random.randint(10, 500, size=1000),  # Additional column for purchase amount
    'Withdrawal_Amount': np.random.randint(10, 500, size=1000),  # Additional column for withdrawal amount
}

# Create a DataFrame from the dummy data
df = pd.DataFrame(data)

# Count the occurrences of each transaction type
transaction_type_counts = df['Transaction_Type'].value_counts()

# Print the counts of each transaction type
print("Transaction Type Counts:")
print(transaction_type_counts)

# Generate a cross-tabulation of transaction type and customer segment with median purchase amount
cross_tab_with_median = pd.crosstab(df['Transaction_Type'], df['Customer_Segment'], values=df['Purchase_Amount'], aggfunc='median')

# Print the cross-tabulation with median purchase amount
print("\nCross-Tabulation of Transaction Type and Customer Segment with Median Purchase Amount:")
print(cross_tab_with_median)
```

<b> 6. Relative Class Frequency

```python
transaction_type_relative_freq = df['Transaction_Type'].value_counts(normalize=True)
print(transaction_type_relative_freq)
```

<b> 7. Cross-Tabulation

```python
# Create a cross-tabulation of transaction types and customer segments
cross_tab = pd.crosstab(df['Transaction_Type'], df['Customer_Segment'])
print(cross_tab)
```

<b> 8. Extending Cross-Tabulation

```python
# Perform cross-tabulation with aggregated values (e.g., median)
cross_tab_with_median = pd.crosstab(df['Transaction_Type'], df['Customer_Segment'], values=df['Purchase_Amount'], aggfunc='median')
print(cross_tab_with_median)
```

<b> <font size = 12, font color ="yellow"> 12. Navigating Data Science: The Art of Hypothesis Generation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Generate dummy data for flight ticket prices and durations
data = {
    'Price': np.random.randint(100, 1000, size=1000),
    'Duration': np.random.randint(1, 10, size=1000),
    'Total_Stops': np.random.randint(0, 3, size=1000),
    'Group': np.random.choice(['A', 'B'], size=1000),
    'Airline': np.random.choice(['A', 'B', 'C', 'D'], size=1000)
}
df = pd.DataFrame(data)
df
# 1. Hypothesis: There is a correlation between the price of flight tickets and the duration of the flight.


```

```python
# 2. Assessing existing knowledge about the dataset
summary_statistics = df.describe()
summary_statistics
```

```python
# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()
correlation_matrix
```

```python
# 3. Checking for spurious correlations
spurious_corr = df[['Price', 'Duration', 'Total_Stops']].corr()
spurious_corr
```

```python
# 4. Assessing the validity of hypotheses
validity_check = df.sample(n=100)
validity_check
```

```python
# 5. Conducting hypothesis testing
group1 = df[df['Group'] == 'A']['Price']
group2 = df[df['Group'] == 'B']['Price']
t_stat, p_value = ttest_ind(group1, group2)
t_stat, p_value
```

```python
# 6. Avoiding data snooping
significance_threshold = 0.05
if p_value < significance_threshold:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")

```

```python
# 7. Hypothesis generation through EDA
sns.barplot(x='Airline', y='Duration', data=df)
plt.title('Mean Flight Duration by Airline')
plt.show()

```

```python
# 8. Designing the experiment
sample_size = 1000
sample_data = df.sample(n=sample_size)
sample_data
```
