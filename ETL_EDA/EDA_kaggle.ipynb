```python
def iqr_f(df, column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return df[(column > lower) & (column < upper)]
```

```python
data = iqr_f(data, data["Height"])
data = iqr_f(data, data["Weight"])
data = iqr_f(data, data["BMI"])
```

##  What Skewness Tells Us

**Skewness** measures how asymmetric a distribution is:

- **0** → perfectly symmetrical (like a normal bell curve)
- **Positive (> 0)** → right-skewed (tail to the right)
- **Negative (< 0)** → left-skewed (tail to the left)

---

## 🟡 When Is Skew Significant?

| Skewness Value Range      | Interpretation            | Do You Transform? |
|---------------------------|----------------------------|--------------------|
| Between **-0.5 and +0.5** | Roughly symmetrical        | ❌ No need         |
| Between **±0.5 and ±1**   | Moderately skewed          | 🤔 Maybe           |
| Beyond **±1**             | Highly skewed              | ✅ Yes (usually)   |

So if your column has a **skewness > +1**, a **log transform** or similar is strongly recommended to make the data more normal-like.

---

## 🔁 What Log Transform Actually Does

Log transformation:
- **Compresses large values**, reducing the spread of the tail
- **Pulls extreme outliers back in** toward the center
- Makes the **distribution more symmetric**, often improving model performance

---

## ⚠️ Don’t Forget These

- Use `np.log1p()` instead of `np.log()` if the data includes **zeros**
- Don’t apply log transform to **negative** values unless you've shifted the data (e.g., added a constant)

2. Categorical (Binary) – Numerical (e.g. Smoking vs. BMI)
Point Biserial Correlation

```python
data.sample(5)
```

```python
from scipy.stats import pointbiserialr

# Make sure Smoking is encoded as binary: 1 = Yes, 0 = No
data['Smoking_binary'] = data['Smoking'].map({'Yes': 1, 'No': 0})

r, p = pointbiserialr(data['Smoking_binary'], data['BMI'])
print(f"Point Biserial r = {r:.3f}, p-value = {p:.3f}")
```

✅ Point Biserial r = 0.014, p = 0.326 There is no statistically significant relationship between Smoking status (Yes/No) and BMI.

Since p > 0.05, we fail to reject the null hypothesis.

Interpretation: Smokers and non-smokers in your data have very similar BMI distributions—no meaningful linear difference detected.

3. Categorical (Multi) – Numerical (e.g. Education Level vs. Weight)

Kruskal-Wallis H-test

```python
from scipy.stats import kruskal

groups = [group["Weight"].values for name, group in data.groupby("Education_level")]
stat, p = kruskal(*groups)
print(f"Kruskal-Wallis H = {stat:.3f}, p-value = {p:.3f}")
```

✅ Kruskal-Wallis H = 63.793, p = 0.000 There is a statistically significant difference in Weight across Education levels.

Since p < 0.05, we reject the null hypothesis.

Interpretation: At least one education group has a different median weight than the others.

4. Categorical – Categorical (e.g. Smoking vs. Gender)

Chi-Square Test + Cramér’s V

```python
from scipy.stats import chi2_contingency
import numpy as np

# Contingency table
ct = pd.crosstab(data["Smoking"], data["Gender"])

# Chi-squared test
chi2, p, dof, expected = chi2_contingency(ct)

# Cramér’s V calculation
n = ct.sum().sum()
phi2 = chi2 / n
r, k = ct.shape
cramer_v = np.sqrt(phi2 / min(k - 1, r - 1))

print(f"Chi² = {chi2:.3f}, p = {p:.3f}, Cramér’s V = {cramer_v:.3f}")
```

✅ Chi² = 259.102, p = 0.000, Cramér’s V = 0.224 There is a statistically significant association between Smoking and Gender.

Since p < 0.05, we reject the null hypothesis of independence.

Interpretation: Smoking behavior differs between genders, and the strength of association (Cramér’s V = 0.224) is small to moderate.

```python
from scipy.stats import ttest_ind

group1 = data[data['Smoking'] == 'Yes']['BMI']
group2 = data[data['Smoking'] == 'No']['BMI']

t_stat, p = ttest_ind(group1, group2)
print(f"T = {t_stat:.3f}, p = {p:.3f}")
```
