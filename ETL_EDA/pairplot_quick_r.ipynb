### üéØ `diag_kind`: What it Controls

The `diag_kind` argument in `sns.pairplot()` controls the **type of plot shown on the diagonal** of the pairplot matrix, where each variable is plotted against itself.

You can choose:

| Option       | Meaning                                           |
|--------------|---------------------------------------------------|
| `'hist'`     | Shows a histogram (default) for each variable     |
| `'kde'`      | Shows a KDE (Kernel Density Estimate) curve       |

---

### ‚ú® Visual Comparison

- `diag_kind='hist'`:
  - Bar-like plots
  - Shows **counts** per bin
  - Great for seeing exact distributions

- `diag_kind='kde'`:
  - Smooth curves
  - Gives a **smoothed estimate** of the distribution
  - Better for detecting distribution shape (e.g., skew, multimodal)

---




### üé® What `palette` Does:
It sets the **color theme** for your plot. You can use:
- Named Seaborn palettes: `"deep"`, `"muted"`, `"pastel"`, `"Set2"`, etc.
- Custom lists of color codes: `["#FF9999", "#66B3FF"]`
- Built-in color maps from **matplotlib**: `"coolwarm"`, `"viridis"`, etc.

### üéØ `plot_kws` ‚Äî Customize Off-Diagonal Scatterplots

This parameter lets you pass a dictionary of keyword arguments to control the **appearance** of the scatter plots (the plots below the diagonal).  

#### Example:
```python
sns.pairplot(data=df, hue="Gender", diag_kind="kde", corner=True,
             plot_kws=dict(alpha=0.1, edgecolor='none'))
```

#### What this does:
- **`alpha=0.1`** ‚Üí Makes each point 90% transparent ‚Üí helps reduce clutter when there are many points  
- **`edgecolor='none'`** ‚Üí Removes the border around each point ‚Üí cleaner look in dense scatterplots

You can also customize:
- `s=` for point size
- `marker=` for shape
- `linewidth=` to style outlines
