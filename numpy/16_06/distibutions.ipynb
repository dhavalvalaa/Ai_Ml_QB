```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(size=100), kind="kde")  # Less smooth
sns.displot(random.normal(size=5000), kind="kde")  # Very smooth

plt.show()
```

```python
from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
sns.displot(x)
```

```python
from scipy.stats import poisson

lambda_ = 10  # Average calls per hour
k = 12        # Desired number of calls
probability = poisson.pmf(k, lambda_)

print(f"Probability of 12 calls in an hour: {probability:.4f}")  
```
