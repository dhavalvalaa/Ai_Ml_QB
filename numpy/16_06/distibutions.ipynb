```python
import numpy as np
import matplotlib.pyplot as plt

# Gaussian Kernel Function
def gaussian_kernel(x, xi, bandwidth):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - xi) / bandwidth) ** 2)

# KDE Function
def kde_manual(x_values, data, bandwidth):
    kde_values = np.zeros_like(x_values)
    n = len(data)
    
    for i, x in enumerate(x_values):
        kde_values[i] = np.sum([gaussian_kernel(x, xi, bandwidth) for xi in data]) / (n * bandwidth)
    
    return kde_values

# Sample Data
data = np.array([30, 45, 60, 75])
x_values = np.linspace(20, 90, 100)  # Points to evaluate KDE
bandwidth = 5  # Adjust bandwidth for smoothness

# Compute KDE
kde_values = kde_manual(x_values, data, bandwidth)



for i in kde_values:
    print(i)

# Plot Results
plt.plot(x_values, kde_values, label="Manual KDE", color="blue")
plt.scatter(data, np.zeros_like(data), color="red", label="Data Points", zorder=3)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Kernel Density Estimation (Manual Calculation)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Step 1: Simulate past normal behavior (e.g., daily spending in $)
past_transactions = np.random.normal(loc=50, scale=10, size=1000)  # average $50, normal behavior

# Step 2: New transaction comes in
new_transaction = int(input("enter transcation amont"))  # the user suddenly spent $120 today

# Step 3: Calculate deviation (Chi-Square score)
mean = np.mean(past_transactions)
std = np.std(past_transactions)

# Chi-Square score (simplified as squared deviation over variance)
chi_square_score = ((new_transaction - mean) ** 2) / (std ** 2)

# Step 4: Set a threshold (e.g., 95% confidence)
threshold = chi2.ppf(0.95, df=1)  # df=1 for one feature

# Step 5: Decision
if chi_square_score > threshold:
    print(f"⚠️ Suspicious transaction! Score = {chi_square_score:.2f} > Threshold = {threshold:.2f}")
else:
    print(f"✅ Normal transaction. Score = {chi_square_score:.2f} ≤ Threshold = {threshold:.2f}")

# (Optional) Visualize
plt.hist(past_transactions, bins=30, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.axvline(new_transaction, color='red', linestyle='dashed', linewidth=2, label="New Transaction")
plt.title("Past User Transactions")
plt.xlabel("Transaction Amount ($)")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import chi2

# -----------------------------
# 1)  Historical “normal” data
# -----------------------------
past = np.array([50, 52, 48, 51, 49, 53, 47, 50, 51, 52])  # we *assume* these are normal

# Robust centre & spread (median & MAD) so a hidden fraud doesn’t skew them
median = np.median(past)
mad    = np.median(np.abs(past - median))           # Median Absolute Deviation
sigma  = 1.4826 * mad                               # Convert MAD to σ‑like scale

# -----------------------------
# 2)  Score a stream of new txns
# -----------------------------
new_txns = [44, 57, 100, 55, 20, 30, 51]        # some normal, one extreme

# Chi‑Square‑style score for each
scores = [((x - median)**2) / (sigma**2) for x in new_txns]

# -----------------------------
# 3)  χ² threshold (no labels!)
# -----------------------------
alpha      = 0.95                  # keep the safest 95 %
threshold  = chi2.ppf(alpha, df=4) # df = number of features

for x, s in zip(new_txns, scores):
    flag = "⚠️  SUSPECT" if s > threshold else "✅ OK"
    print(f"Txn ${x:>4}  •  score={s:6.2f}  •  {flag}")
```

```python
import numpy as np
from scipy.stats import chi2
from scipy.spatial.distance import mahalanobis
import matplotlib.pyplot as plt

# -----------------------------
# 1. Generate Past Normal Data
# -----------------------------
np.random.seed(42)

# Simulate 100 past transactions, 4 features
# Format: [amount, hour, location_code, merchant_code]
past_data = np.column_stack([
    np.random.normal(50, 5, size=100),       # amount
    np.random.normal(14, 2, size=100),       # hour
    np.random.randint(1, 6, size=100),       # location_code
    np.random.randint(1, 11, size=100),      # merchant_code
])

# Compute mean & inverse covariance matrix
mean_vector = np.mean(past_data, axis=0)
cov_matrix = np.cov(past_data.T)
inv_cov = np.linalg.inv(cov_matrix)

# -----------------------------
# 2. New Incoming Transaction
# -----------------------------
# Normal-ish transaction
txn1 = [52, 15, 3, 5]

# Suspicious transaction (amount is huge, weird hour, odd location)
txn2 = [120, 3, 5, 10]

# -----------------------------
# 3. Mahalanobis Distance & Chi-Square Test
# -----------------------------
def chi_square_score(x, mean, inv_cov):
    return mahalanobis(x, mean, inv_cov) ** 2

# Compute scores
score1 = chi_square_score(txn1, mean_vector, inv_cov)
score2 = chi_square_score(txn2, mean_vector, inv_cov)

# Threshold for df=4
threshold = chi2.ppf(0.95, df=4)

# -----------------------------
# 4. Results
# -----------------------------
print(f"Txn 1 score = {score1:.2f}  --> {'OK ✅' if score1 < threshold else 'SUSPICIOUS ⚠️'}")
print(f"Txn 2 score = {score2:.2f}  --> {'OK ✅' if score2 < threshold else 'SUSPICIOUS ⚠️'}")
print(f"Chi-Square threshold (df=4, 95%) = {threshold:.2f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate transaction data
np.random.seed(0)
amounts = np.random.normal(loc=1000, scale=200, size=1000)  # mean=1000, std=200

# Plot the distribution
plt.hist(amounts, bins=30, density=True, alpha=0.6, color='skyblue')

# Plot the theoretical normal curve
x = np.linspace(400, 1600, 100)
plt.plot(x, norm.pdf(x, 1000, 200), 'r-', label='Normal PDF')

# Add a suspicious threshold line
plt.axvline(1000 + 3*200, color='red', linestyle='--', label='Suspicious Threshold')

plt.title("Normal Distribution of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Density")
plt.legend()
plt.show()
```
