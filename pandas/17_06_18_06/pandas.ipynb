#### 6.2 Replace Empty Values
- Another way of dealing with empty cells is to insert a new value instead.

- This way we do not have to delete entire rows just because of some empty cells.

- The **fillna()** method allows us to replace empty cells with a value

##### 6.2.1 Replacing by a Static value

``````````python
# df.fillna({'Engine Size' : '1.5l'}, inplace=True)
df.fillna({'Price (USD)' : 16000.0}, inplace=True)
df.fillna({'Brand' : 'Honda'}, inplace=True)
df.fillna({'Mileage (km)' : 70000.0 }, inplace=True)

#2nd method but gives the warning because indirectly it uses uper one 
df['Engine Size'].fillna('1.51', inplace=True)
print(df.to_string())
``````````

##### 6.2.2 Replace Using Mean, Median, or Mode

- A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

- Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column

Mean = the average value (the sum of all values divided by number of values).

Median = the value in the middle, after you have sorted all values ascending.

Mode = the value that appears most frequently.

``````````python
print(df["Brand"].mode()) #1st
print(df.Brand.mode()) #2nd

print("-"*40)

print(df["Brand"].mode()[0]) #1st
print(df.Brand.mode()[0]) #2nd

#we uses mode()[0] 0 for the first value (exa - honda)
``````````

``````````python
x = df["Brand"].mode()[0]
y = df["Engine Size"].mode()[0]
z = df["Mileage (km)"].mean()
w = df["Price (USD)"].median()

df.fillna({"Brand": x}, inplace=True)
df.fillna({"Engine Size": y}, inplace=True)
df.fillna({"Mileage (km)": z}, inplace=True)
df.fillna({"Price (USD)": w}, inplace=True)
# df["Brand"].fillna(x, inplace = True)


print(df.to_string())
``````````

### 8. Pandas - Data Correlations

- The corr() method calculates the relationship between each column in your data set.


#### **corr()** Result Explained

- The Result of the corr() method is a table with a lot of numbers that represents how well the relationship is between two columns.

- The number varies from -1 to 1.

1. Perfect Correlation 
    - 1 means that there is a 1 to 1 relationship.
    - And for this data set, each time a value went up in the first column, the other one went up as well.

2. Good Correlation 
    - 0.9 is also a good relationship.
    - And if you increase one value, the other will probably increase as well.
    - -0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

3. Bad Correlation
    - 0.2 means NOT a good relationship.
    - meaning that if one value goes up does not mean that the other will.

``````````python

# Method	    Best For	                        Why Use?
# Pearson	Linear relationship	        Precise for normal, proportionally growing data.
# Kendall	Ordinal/ranked data	        Works with rankings and handles tied values.
# Spearman	Monotonic relationship	    Useful when data changes consistently but not linearly.
``````````

``````````python
print(df.corr(numeric_only=True, method='kendall'))
print(df.corr(numeric_only=True))
``````````

``````````python
# 1. Pearson Correlation (Linear Relationship)
# Use case: Stock market trends – Analyzing the correlation between a company's advertising spend and its revenue.

import pandas as pd

# Sample Data
data = {
    "Advertising_Spend": [10, 20, 30, 40, 50],
    "Revenue": [100, 200, 300, 400, 500]
}

df = pd.DataFrame(data)

# Compute Pearson Correlation
print(df.corr(method="pearson"))


# 2. Kendall Correlation (Ordinal Association)
# Use case: Employee performance – Ranking employees by years of experience vs. promotion frequency.

data = {
    "Years_of_Experience": [1, 3, 5, 7, 10],
    "Promotions": [0, 1, 1, 2, 3]
}

df = pd.DataFrame(data)

# Compute Kendall Correlation
print(df.corr(method="kendall"))


# 3. Spearman Correlation (Monotonic Relationship)
# Use case: Customer satisfaction – Analyzing product ratings vs. likelihood of repeat purchases.

data = {
    "Product_Rating": [5, 4, 3, 2, 1],
    "Repeat_Purchases": [90, 70, 40, 20, 5]
}

df = pd.DataFrame(data)

# Compute Spearman Correlation
print(df.corr(method="spearman"))
``````````

#### The groupby() function in Pandas involves three main steps: Splitting, Applying, and Combining.


##### Splitting: This step involves dividing the DataFrame into groups based on some criteria. The groups are defined by unique values in one or more columns.
##### Applying: In this step, a function is applied to each group independently. You can apply various functions to each group, such as:
##### - Aggregation: Calculate summary statistics (e.g., sum, mean, count) for each group.
##### - Transformation: Modify the values within each group.
##### - Filtering: Keep or discard groups based on certain conditions.
##### Combining: Finally, the results of the applied function are combined into a new DataFrame or Series.
